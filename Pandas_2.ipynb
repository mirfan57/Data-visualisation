```python
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2, on='key')  # this is by defult inner merge with only intersecting data
```

```python
# Left merge will consider everything from my left dataframe and whatever data is not present in right table will be NaN.
pd.merge(df1,df2, on='key', how='left')
```

```python
pd.merge(df1,df2, on='key', how="right")
```

```python
# Outer join will consider both left and right data frames completely & wherever value not present will be filled with NaN.
pd.merge(df1,df2, on='key', how="outer")
```

### JOIN OPERATIONS

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

df2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
df1
```

```python
df2
```

```python
# join operation performs join based on indexes while merge was based on columns.
df1.join(df2)    # similar to left join where df1 is left dataframe and df2 is right.
# by default it takes how='left' which we can change to right.
```

```python
df1.join(df2, how='right')  # here df2 is my right dataframe.
```

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(20)
df = pd.DataFrame(np.random.randn(10,4), index = pd.date_range('2022-06-28', periods=10), columns =['a','b','c','d'])
```

```python
df
```

```python
df['new'] = df['c'] + df['d']
```

```python
df
```

```python
df['square_col_d'] = df['d'].apply(lambda x: x*x)
df
```

```python
df['log_col_d'] = df['d'].apply(lambda x: np.log(x))
df
```

If there are negative data in column d, then keep it as it is and replace other data by their respective log values in new_col_d.

```python
# Define function to keep negative data untouched and log of remaining.
def test(x):
    if (x<0):
        return x
    else:
        return np.log(x)
```

```python
df['new_log_col_d'] = df['d'].apply(test)
```

```python
df
```

```python
If dataset in col a is positive then 1 & if negative then 0, print in new column.
```

```python
def test1(x):
    if x<0:
        return 1
    else:
        return 0
```

```python
df['bin_col_a'] = df['a'].apply(test1)
```

```python
df
```
