```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# relationship plot
# by default knid = 'scatter'
sns.relplot(x='tip', y= 'total_bill', data=tips)
```

```python
# Hue: grouping of given column based on color
# I have to seperate smoker and non-smoker by different colors on same graph.
sns.relplot(x='tip', y= 'total_bill', data=tips, hue='smoker')
```

```python
sns.relplot(x='tip', y= 'total_bill', data=tips, hue='day')
```

```python
# Style: grouping of given column based on different shape/pattern.
# On week days, we want how many were smokers & NON-SMOKERS.
sns.relplot(x='tip', y= 'total_bill', data=tips, hue='day', style='smoker')
```

```python
sns.relplot(x='tip', y= 'total_bill', data=tips, hue='day', style='smoker', size='size')
```

says if we gonna collect enough amount of samples from any distribution, then for sure collection of sample is going to showcase a normal distribution. Sample is going to represent your entire population of data and showcase central tendency of data.
