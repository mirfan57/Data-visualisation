```python
import numpy as np
import seaborn as sns
```

### Standard normal distribution

```python
np.random.normal(0, 1, 100)
```

```python
import statsmodels.api as sm
import pylab as py

# np.random generates different random numbers whenever the code is executed
# Note: When you execute the same code the graph look different than shown below.

# Random data points generated
data_points = np.random.normal(0, 1, 100)	

sm.qqplot(data_points, line ='45')
py.show()

# sm plot will initialize some std. normal distribution data points and plot on x-axis vs our data_points on y-axis. If the 
# resultant points fall on a 45 degree line, then we can say that our data_points are gaussian distributed.
```

```python
sns.histplot(data_points)
```

```python
sns.displot(data_points, kind="kde")
```

### Create Log Normal Distributed data

```python
# import numpy
import numpy as np
  
# using np.lognormal() method
gfg = np.random.lognormal(0, 1, 100)

print(gfg) 
```

```python
sns.histplot(gfg)
```

```python
sns.displot(gfg, kind="kde")
```

```python
# Taking log of data points of lognormal plot should give a gaussian distribution.
np.log(gfg)
```

```python
# Q-Q Plot
sm.qqplot(np.log(gfg), line ='45')
py.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt

# generate non-normal data (exponential)
original_data = np.random.exponential(size=1000)

#transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(original_data)

#creating axes to draw plots
fig, ax = plt.subplots(1,2)

#plotting the original (non-normal) & fitted (normal) data
sns.distplot(original_data, hist=False, kde=True,
            kde_kws = {'shade':True, 'linewidth':2},
            label="Non-Normal", color = "green", ax= ax[0])

sns.distplot(fitted_data, hist=False, kde=True,
            kde_kws = {'shade':True, 'linewidth':2},
            label="Normal", color = "green", ax= ax[1])

#adding legends to the sub-plot
plt.legend(loc="upper right")

#rescaling the subplots
fig.set_figheight(5)
fig.set_figwidth(10)

print(f"Lambda value used for transformation: {fitted_lambda}")
```

```python
sm.qqplot(original_data, line ='45')
py.show()       # not gaussian as blue dots don't lie on the 45 degree line.
```

```python
sm.qqplot(np.log(original_data), line ='45')
py.show()
```

```python
sm.qqplot(fitted_data, line ='45')
py.show()      # here, data after boxcox transformation is also not gaussian as the points not completely lie on the red line.
```

### After applying boxcox transformation we apply q-q plot to check. So, if tramsformed data is normaly distribited then we can say that original data is pareto distribution.

```python
import scipy.stats as stats
import pylab

x_m = 10
alpha = 15    # height
size = 100000
samples = (np.random.pareto(alpha, size) + 1) * x_m  # x_m is multiplied to make the scale bigger, it is just noise
stats.probplot(samples, dist='pareto', sparams = (15,10), plot=pylab) # we have to compare with pareto dist. probplot is like q-q plot
plt.show
```

```python
samples = (np.random.pareto(alpha, size) + 1) * x_m
sns.displot(samples, kind="kde")
```

```python
samples = np.random.normal(0,1,100)
stats.probplot(samples, dist='norm', sparams = (15,10), plot=pylab) 
plt.show
# as points are falling on the line, we can say that samples is normally distributed.
```
